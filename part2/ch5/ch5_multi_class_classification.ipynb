# 2. 라이브러리 및 데이터 불러오기

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd

# train = pd.read_csv("train.csv")
# test = pd.read_csv("test.csv")
train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part2/ch5/train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part2/ch5/test.csv")
```

```python
# train.shape, test.shape

train.describe(include='O')
test.describe(include='O')
# Payment_of_Min_Amount, Credit_Mix, Payment_Behaviour
```

```python
test.isnull().sum()
```

```python
y_train = train.pop('Credit_Score')
train.shape
```

```python
# 인코딩
# Payment_of_Min_Amount, Credit_Mix, Payment_Behaviour
from sklearn.preprocessing import LabelEncoder
cols = ['Payment_of_Min_Amount', 'Credit_Mix', 'Payment_Behaviour']
# df = pd.concat([train, test])
for col in cols:
  le = LabelEncoder()
  # df[col] = le.fit_transform(df[col])
  train[col] = le.fit_transform(train[col])
  test[col] = le.transform(test[col])
# train = df.iloc[:len(train)]
# test = df.iloc[len(train):]
train
```

```python
# missing, outliner, encoding, scale
train[cols].info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(train, y_train, test_size=0.2, random_state=0)
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
# 학습
from sklearn.ensemble import RandomForestClassifier
import lightgbm as lgbm

rf = RandomForestClassifier(random_state=0)
lgbmc = lgbm.LGBMClassifier(random_state=0)

rf.fit(X_train, y_train)
lgbmc.fit(X_train, y_train)
```

```python
pred_rf = rf.predict(X_val)
pred_lgbmc = lgbmc.predict(X_val)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
ac_rf = accuracy_score(y_val, pred_rf)
f1_rf = f1_score(y_val, pred_rf, average='macro')

ac_lg = accuracy_score(y_val, pred_lgbmc)
f1_lg = f1_score(y_val, pred_lgbmc, average='macro')

print(ac_rf, f1_rf)
print(ac_lg, f1_lg)
# rf win
```

```python
pred = rf.predict(test)
submit = pd.DataFrame({'pred': pred})
submit.to_csv('result.csv', index=False)
pd.read_csv('result.csv')
```

# 4. 데이터 전처리

```python
target = train.pop('Credit_Score')
cols = train.columns[train.dtypes == object]
cols
```

```python
train['Payment_of_Min_Amount'] = train['Payment_of_Min_Amount'].astype('category')
train['Credit_Mix'] = train['Credit_Mix'].astype('category')
train['Payment_Behaviour'] = train['Payment_Behaviour'].astype('category')

test['Payment_of_Min_Amount'] = test['Payment_of_Min_Amount'].astype('category')
test['Credit_Mix'] = test['Credit_Mix'].astype('category')
test['Payment_Behaviour'] = test['Payment_Behaviour'].astype('category')
```

# 5. 검증 데이터 분할

```python
# 검증 데이터 나누기
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(
    train,
    target,
    test_size=0.2,
    random_state=0)
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

# 6. 머신러닝 학습 및 평가

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
# 랜덤포레스트(에러)
# from sklearn.ensemble import RandomForestClassifier
# rf = RandomForestClassifier(random_state=0)
# rf.fit(X_train, y_train)
```

```python
# LightGBM
import lightgbm as lgb
lgbmc = lgb.LGBMClassifier(random_state=0, verbose=-1)
lgbmc.fit(X_train, y_train)
pred = lgbmc.predict(X_val)
pred
```

```python
accuracy = accuracy_score(y_val, pred)
print('accuracy_score:', accuracy)

f1 = f1_score(y_val, pred, average='macro')
print('f1_score:', f1)
```
