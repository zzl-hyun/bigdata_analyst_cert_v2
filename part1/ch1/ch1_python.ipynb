### 리턴이 있는 함수

```python
# 함수 정의 (리턴 값)
def plus(x, y):
    result = x+y
    return result

# 함수 호출
a = plus(2,3)
print(a)
```

```python
# 리턴값이 2개 (최소, 최대값을 구하는 함수)
listbox = [15, 46, 78, 24, 56]
def min_max(data):
    mi = min(data)
    ma = max(data)
    return mi, ma

a, b = min_max(listbox)
print(a,b)
```

```python
# 평균을 구하는 함수
listbox = [15, 46, 78, 24, 56]

def mean(data):
    return sum(data) / len(data)

mean(listbox)
```

```python
# Q9. 리스트에서 최대값에서 최소값을 뺀 결과를 리턴하는 함수를 만들고 실행하시오
listbox = [15, 46, 78, 24, 56]
def maxmin(data):
    return max(data) - min(data)

maxmin(listbox)
```
