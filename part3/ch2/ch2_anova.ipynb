```python
import pandas as pd
df = pd.DataFrame({
    'A': [10.5, 11.3, 10.8, 9.6, 11.1, 10.2, 10.9, 11.4, 10.5, 10.3],
    'B': [11.9, 12.4, 12.1, 13.2, 12.5, 11.8, 12.2, 12.9, 12.4, 12.3],
    'C': [11.2, 11.7, 11.6, 10.9, 11.3, 11.1, 10.8, 11.5, 11.4, 11.0],
    'D': [9.8, 9.4, 9.1, 9.5, 9.6, 9.9, 9.2, 9.7, 9.3, 9.4]
})
print(df.head(2))
# 주어진 데이터는 4종류의 비료를 사용한 식물의 성장에 대한 실험 결과다.
# 이 실험에서는 비슷한 조건의 식물 40개를 무작위로 10개씩 나누고 화학 비료 A, B, C, D를 일정 기간 사용한 후 성장량을 측정했다.
# 성장의 차이가 있는지 유의수준 0.05하에서 검정하시오

# 귀무가설: 네 가지 비료의 효과는 동일하다
# 대립가설: 비료의 효과에는 차이가 있다.

# 1. 정규성 검정 -> 등분산성 검정
from scipy import stats

# 정규성 shapiro
print(stats.shapiro(df['A']))
print(stats.shapiro(df['B']))
print(stats.shapiro(df['C']))
print(stats.shapiro(df['D']))
# 0.05보다 크니 정규성 만족 -> 등분산성 levene
print(stats.levene(df['A'],df['B'],df['C'],df['D']))
# 0.05보다 크니 등분산성 만족

# 일원 분산분석
print(stats.f_oneway(df['A'],df['B'],df['C'],df['D']))
# 0.05보다 작으니 대립가설 채택 -> 차이가 있다!

## 정규성, 등분산 만족하지 않는 케이스는 안다룬다 -> 시험에서 안나온다 -> 그럼 그냥 할필요없다 -> f_oneway만 보면 된다?
```

```python
from scipy import stats

print("=== 정규성 검정 ===")
print(stats.shapiro(df['A']))
print(stats.shapiro(df['B']))
print(stats.shapiro(df['C']))
print(stats.shapiro(df['D']))

print("\n === 등분산 검정 ===")
print(stats.levene(df['A'], df['B'], df['C'], df['D']))

print("\n === 일원 분산 분석 ===")
print(stats.f_oneway(df['A'], df['B'], df['C'], df['D']))
```

# 이원 분산 분석

```python
# 데이터는 네 가지 종류의 나무에 대해 세 가지 종류의 비료를 사용해 성장률을 조사한 결과다.
# 비료 간 및 나무 종류 간의 성장률 차이가 있는지 유의수준 0.05하에서 검정하시오

import pandas as pd
# df = pd.read_csv("tree.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part3/ch2/tree.csv")

from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

model = ols('성장률 ~ 나무 + 비료 + 나무:비료',df).fit()
print(anova_lm(model))
```

```python
import pandas as pd
# df = pd.read_csv("tree.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part3/ch2/tree.csv")
# print(df.sample(10))
df.
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('성장률 ~ C(나무) + C(비료) + C(나무):C(비료)', data=df).fit()
anova_table = sm.stats.anova_lm(model)
print(anova_table)
```
