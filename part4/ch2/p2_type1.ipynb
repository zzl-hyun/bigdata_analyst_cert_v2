```python
# 1. views를 ㅣㄱ준으로 내림차순 정렬하고
# 상위 10개 데이터를 상위 10번째의 값으로 대체해서
# age가 80이상인 데이터의 view 평균을 구해라
```

```python
import pandas as pd
# df = pd.read_csv("members.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch2/members.csv")

index = df['views'].nlargest(10).index
var = df['views'].nlargest(10).iloc[9]
df.loc[index, 'views'] = var

df[df['age'] >= 80]['views'].mean().round(2)
```

```python
import pandas as pd
# df = pd.read_csv("members.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch2/members.csv")

# 1) views 컬럼 내림차순 정렬
df.sort_values('views', ascending=False, inplace=True)
# 2) 상위 10개 중에서 10번째 (최소값) 값 구하기
min_value = df['views'][:10].min()

# 3) 상위 10개 중에서 10번째 (최소값) 값 대체
df.iloc[:10,-1] = min_value

# 4) age가 80 이상의 views 컬럼 평균
cond = df['age'] >= 80
result = df[cond]['views'].mean()

# 5) 반올림하여 소수 둘째자리까지 계산
print(round(result, 2))
```

### 문제2

```python
# 앞에서부터 80%만 활용해 f1의 결측치를 중앙값으로 채우고 전후 표본표준편차를 구하고 두 표본 표준편차의 차이를 계산하시오(절대값, 소수 둘쨰자리까지)
import pandas as pd
# df = pd.read_csv("members.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch2/members.csv")
df = df.iloc[:int(len(df)*0.8)]

before = df
after = df.copy()
after['f1'] = after['f1'].fillna(after['f1'].median())

x = df['f1'].std()
y = after['f1'].std()
print(round(abs(x-y),2))
```

### 문제3.

```python
## age의 이상치를 모두 더해라
## 이상치는 평균으로부터 표준편차 * 1.5를 벗어나는 영역

import pandas as pd
# df = pd.read_csv("members.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch2/members.csv")

avg = df['age'].mean()
std = df['age'].std()

upper = avg + 1.5*std
lower = avg - 1.5*std
x = df[df['age'] > upper]['age'].sum()
y = df[df['age'] < lower]['age'].sum()
print(x + y)
```
